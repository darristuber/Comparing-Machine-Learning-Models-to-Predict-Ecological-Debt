#Background:


##What is an Ecological Footprint?

Ecological footprint is a measure of how much land and water area is required to produce the resources and services consumed by an individual, population, or activity. It is expressed in terms of global hectares (gha), which is a standardized measure that accounts for the differing productivity of land and water resources in different regions of the world.

The ecological footprint includes two components: the demand on natural resources and the capacity of ecosystems to regenerate those resources. The demand on natural resources includes the consumption of food, fiber, timber, and energy, as well as the use of land for buildings and infrastructure. The capacity of ecosystems to regenerate those resources includes the ability of land and water resources to produce food, fiber, timber, and other resources, as well as the capacity of forests, oceans, and other ecosystems to absorb carbon dioxide and other pollutants.

The ecological footprint is calculated by summing up the area of land and water that is required to produce the resources and services consumed by an individual, population, or activity, and then comparing that with the available land and water resources. If the ecological footprint exceeds the available land and water resources, it indicates that the consumption of natural resources is unsustainable.

The ecological footprint concept was first introduced in the early 1990s by Mathis Wackernagel and William Rees as a way of quantifying the environmental impact of human activities. Today, the ecological footprint is widely used as a tool for environmental accounting and sustainability assessment, and it has been applied to a range of activities, including individual lifestyles, businesses, and entire countries. Environmental footprint is becoming a more prominent factor in policy as well, with studies indicating that it is a more comprehensive  measurement of environmental degration than carbon footprint, and thus can promote more effective and long lasting policies (Sarkodie, 2021). Furthermore, by analyzing a countries ecological footprint and biocapacity, patterns can be identified to assure the most effective policy strategies (2012 ScienceDirect).

###Our Data Set:
We retrieved this dataset from kaggle (Devastator, 2022), the original source was the 2018 Edition of the National Footprint Accounts from the Global Footprint Network (Evans, 2017). It discusses the various impacts that crop land, grazing land, forest land, fishing ground, and built-up land have on the ecological footprint of a country and biocapacity of a country.
  This project will primarily examine the ecological footprint of consumption per capita, which is the area per person needed to produce the materials consumed by a population and absorb the carbon dioxide emissions associated with the population's consumption. The units of area used are global hectares (gha), or biologically productive metric units of land. The data also offers insights into each nation's biocapacity, or the measure (in gha)  of the land's ability to regenerate after use (by producing additional resources or absorbing waste. If a country's biocapacity is lower than its ecological footprint, it is said to have an ecological deficit, or ecological debtor. Otherwise, it may have an ecological reserve, or be an ecological creditor. These data scores are calculated from input data from the UN Food and Agriculture Organization - and reviewed by researchers at the Global Footprint Network (Lin et al, 2018). There have been some ciritisisms of the inconsistencies found in the GFN method analysis of an environmental footprint using this UN data, though it functions similarily to alternate methods of evaluation such as the hybrid myltiregional input-output model method - where EF scores are formulated using national greenhouse gas emissions (Weinzettel, 2014)
Quality scores of the data are assigned based on the reliability of the results (Global Footprint Network, 2021). Deciphering this data allows us to better communicate the causes of environmental issues and allows them to be addressed in future policy and scientific research. For example, previous studies have supported the "scale effects hypothesis", where "economic development is characterized by natural resource exploitation leading to environmental degradation, a situation that has global policy implications"(Sarkodie, 2021). Through our analysis, we can better understand the source of environmental destructions and how to address them on a national and global scale. This aquisition of knowledge is necessary to implement solutions that can better the world of mankind (Kitzes, 2008).


###**Vocab**:
- **Global hectare** (gha): biologically productive metric area unit of land
- **Ecological Footprint**: Measure (in gha) of how much area an entity needs to produce the resources it consumes and absorb the waste it generates
  - **Consumption**: EF defined by consumption of biocapacity  by a countries inhabitants

    - “the area used to support a defined population’s consumption. The consumption Footprint (in gha) includes the area needed to produce the materials consumed and the area needed to absorb the carbon dioxide emissions. The consumption Footprint of a nation is calculated in the National Footprint and Biocapacity Accounts as a nation’s primary production Footprint plus the Footprint of imports minus the Footprint of exports, and is thus, strictly speaking, a Footprint of apparent consumption.” (source)

- **Production**: EF defined by consumption of biocapacity resulting from production processes
    - “the sum of the Footprints for all of the resources harvested and all of the waste generated within the defined geographical region” (source)
- **Biocapacity**: Measure (in gha) of the lands ability to regenerate after use (to produce resources or absorb waste)
